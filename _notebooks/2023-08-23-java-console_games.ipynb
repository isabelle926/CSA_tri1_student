````java
// simplifying the code
import java.util.Scanner;
import java.lang.Math;
import java.util.Random;

public class ConsoleGame {
    private static final String DEFAULT = "\u001B[0m";

    public ConsoleGame() {
        Scanner sc = new Scanner(System.in);
        boolean quit = false;

        while (!quit) {
            printMenu();
            int choice = getUserChoice(sc);
            quit = performAction(choice);
        }

        sc.close();
    }

    public static void main(String[] args) {
        new ConsoleGame();
    }

    private void printMenu() {
        String menuText = """
                \u001B[35m___________________________
                |~~~~~~~~~~~~~~~~~~~~~~~~~|
                |\u001B[0m          Menu!          \u001B[35m|
                |~~~~~~~~~~~~~~~~~~~~~~~~~|
                | 0 - Exit                |
                | 1 - Rock Paper Scissors |
                | 2 - Higher or Lower     |
                | 3 - Tic Tac Toe         |
                |_________________________|   \u001B[0m

                Choose an option.
                """;
        System.out.println(menuText);
    }

    private int getUserChoice(Scanner sc) {
        System.out.print("Enter your choice: ");
        try {
            return sc.nextInt();
        } catch (Exception e) {
            sc.nextLine();
            System.out.println(e + ": Not a number, try again.");
            return getUserChoice(sc);
        }
    }

    private boolean performAction(int selection) {
        switch (selection) {
            case 0:
                System.out.println("Goodbye, World!");
                return true;
            case 1:
                playRockPaperScissors();
                break;
            case 2:
                playHigherOrLower();
                break;
            case 3:
                playTicTacToe();
                break;
            default:
                System.out.println("Unexpected choice, try again.");
        }
        System.out.println(DEFAULT);
        return false;
    }

    private void playHigherOrLower() {
        System.out.println("Higher or Lower");
        System.out.println("You have three guesses to guess the number I am thinking of between 1-8.");

        Scanner scHL = new Scanner(System.in);
        int randomG = (int) (Math.random() * 8) + 1;

        for (int i = 3; i > 0; i--) {
            int guess = scHL.nextInt();
            if (guess == randomG) {
                System.out.println("You win!");
                break;
            } else if (guess > randomG) {
                System.out.println("The number is lower");
            } else if (guess < randomG) {
                System.out.println("The number is higher");
            }
        }

        System.out.println("Game over.");
        scHL.close();
    }

    private void playRockPaperScissors() {
        System.out.println("Rock Paper Scissors");
        System.out.println("Type r for rock, p for paper, or s for scissors");
    
        Scanner scRPS = new Scanner(System.in);
        String userChoice = scRPS.nextLine().toLowerCase();
        int random = (int) (Math.random() * 3);
        
        String[] choices = {"rock", "paper", "scissors"};
        String userSelection = "";
        String computerSelection = choices[random];
    
        if (userChoice.equals("r")) userSelection = "rock";
        else if (userChoice.equals("p")) userSelection = "paper";
        else if (userChoice.equals("s")) userSelection = "scissors";
        else {
            System.out.println("Invalid input, try again");
            scRPS.close();
            return;
        }
    
        System.out.println("You chose " + userSelection);
        System.out.println("The computer chose " + computerSelection);
    
        if (userSelection.equals(computerSelection)) {
            System.out.println("It's a tie!");
        } else if ((userSelection.equals("rock") && computerSelection.equals("scissors")) ||
                   (userSelection.equals("paper") && computerSelection.equals("rock")) ||
                   (userSelection.equals("scissors") && computerSelection.equals("paper"))) {
            System.out.println("You win!");
        } else {
            System.out.println("You lose!");
        }
    
        scRPS.close();
    }
 
    private void playTicTacToe() {
        //instance variables. char is a data type that is used to store a single character
        /*
        private char[][] board; //2D char array that represents the game board.
        private char currentPlayer; //whether the player is X or O
        private Map<Character, String> players; //makes letter and player a key and value pair. ex: X, Player 1

        
        public playTicTacToe() {
            board = new char[3][3]; //3x3 board
            currentPlayer = 'X'; //sets player to X
            players = new HashMap<>();
            //accessing this data with HashMap
            players.put('X', "Player 1"); //equates X to player 1
            players.put('O', "Player 2"); //equates O to player 2
            initializeBoard(); //clear board
        }
    
        public void play() {
            Scanner scanner = new Scanner(System.in);
            System.out.println("Welcome to Tic Tac Toe!");
    
            while (true) {
                //generate board
                displayBoard();
    
                //player is prompted to enter a row and column for their move, and the input is validated (see if theres alr an X or O there)
                int row, col;
    
                do {
                    System.out.print(players.get(currentPlayer) + ", enter row (1-3) and column (1-3): ");
                    row = scanner.nextInt() - 1;
                    col = scanner.nextInt() - 1;
                } while (!isValidMove(row, col));
    
                //takes values that player entered and uses it to validate move, put move on the board, etc
                makeMove(row, col);
                
                //running checkWin and checkDraw functions and outputs result
                if (checkWin()) {
                    displayBoard();
                    System.out.println(players.get(currentPlayer) + " wins!");
                    break;
                } else if (checkDraw()) {
                    displayBoard();
                    System.out.println("It's a draw!");
                    break;
                }
    
                currentPlayer = (currentPlayer == 'X') ? 'O' : 'X';
            }
    
            scanner.close();
        }
    
        //clears the board each game
        private void initializeBoard() {
            for (int i = 0; i < 3; i++) { //rows
                for (int j = 0; j < 3; j++) { //columns
                    board[i][j] = '-'; //fill each with this
                }
            }
        }
    
        //what the board looks like
        private void displayBoard() {
            System.out.println("-------------");
            for (int i = 0; i < 3; i++) {
                System.out.print("| ");
                for (int j = 0; j < 3; j++) {
                    System.out.print(board[i][j] + " | ");
                }
                System.out.println();
                System.out.println("-------------");
            }
        }
    
        private boolean isValidMove(int row, int col) {
            if (row < 0 || row >= 3 || col < 0 || col >= 3) {
                System.out.println("Invalid move. Row and column must be between 1 and 3.");
                return false;
            }
            if (board[row][col] != '-') {
                System.out.println("Invalid move. Cell is already occupied.");
                return false;
            }
            return true;
        }
    
        //places X or O in right spot
        private void makeMove(int row, int col) {
            board[row][col] = currentPlayer;
        }
    
        //checking all combinations of possible wins
        private boolean checkWin() {
            //changing row index, checks for vertical wins
            for (int i = 0; i < 3; i++) { 
                if (board[i][0] == currentPlayer && board[i][1] == currentPlayer && board[i][2] == currentPlayer) {
                    return true;
                }
                //changing col index each time, checks for horizontal wins
                if (board[0][i] == currentPlayer && board[1][i] == currentPlayer && board[2][i] == currentPlayer) {
                    return true;
                }
            }
    
            //check row, col for diagonal wins
            if (board[0][0] == currentPlayer && board[1][1] == currentPlayer && board[2][2] == currentPlayer) {
                return true;
            }
            if (board[0][2] == currentPlayer && board[1][1] == currentPlayer && board[2][0] == currentPlayer) {
                return true;
            }
            return false;
        }
    
        private boolean checkDraw() {
            for (int i = 0; i < 3; i++) {
                for (int j = 0; j < 3; j++) {
                    if (board[i][j] == '-') {
                        return false; // if an empty cell found, game is not a draw
                    }
                }
            }
            return true; // if all cells are filled, game is a draw
        }
    
        public static void main(String[] args) {
            ticTacToe game = new ticTacToe();
            game.play();
        }
        */
    }
}
ConsoleGame.main(null);
````
